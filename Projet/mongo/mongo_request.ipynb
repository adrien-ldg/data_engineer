```python
import pymongo
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
print(client.list_database_names())
```

```python
database = client.basket
print(type(database))
collection = database['nba_player']
```

```python
database.list_collection_names()
```

```python
cur = collection.find({"team":"Boston Celtics"})
for document in cur :
    print('-----')
    print(document)
```

```python
print(collection.distinct('team'))
len(collection.distinct('team'))
```

```python
cur = collection.find().sort([("pts", -1)]).limit(5)
list(cur)
```

```python
cur = collection.find().sort([("pd", -1)]).limit(5)
list(cur)
```

```python
cur = collection.find().sort([("rb", -1)]).limit(5)
list(cur)
```

```python
cur = collection.aggregate([{'$group': {'_id': '$team', 'average_pts': {'$avg': '$pts'}}},
                            {'$sort': {'average_pts': -1}}])
list(cur)
```

```python
cur = collection.aggregate([{'$group': {'_id': '$team', 'nb': {'$sum': 1}}},
                           {'$sort': {'nb': -1}}])
list(cur)
```

```python
cur = collection.aggregate([
    {'$match': {'pts': {'$gte': 20.0}}},
    {'$sort': {'pts': -1}}])

list(cur)
```

```python
import matplotlib.pyplot as plt

results = collection.aggregate([
    {'$match': {'pts': {'$gte': 20.0}}}])

categories = []
counts = []

for result in results:
    categories.append(result["player"])
    counts.append(result["pts"])

# Cr√©ation d'un barplot
plt.figure(figsize=(12, 6))
plt.bar(categories, counts, color='blue')
plt.title("Barplot des joueurs ayant plus de 20pts")
plt.xlabel("player")
plt.ylabel("pts")
plt.xticks(rotation=45, ha="right", rotation_mode="anchor")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

results = collection.aggregate([
    {'$match': {'pts': {'$gte': 20.0}}}])

categories = []
counts = []

for result in results:
    categories.append(result["player"])
    counts.append(result["pd"] + result["pts"] + result["rb"])

# Cr√©ation d'un barplot
plt.figure(figsize=(12, 6))
plt.bar(categories, counts, color='blue')
plt.title("Barplot des joueurs ayant plus de 20pts")
plt.xlabel("player")
plt.ylabel("points + passe d√©cisive + rebond")
plt.xticks(rotation=45, ha="right", rotation_mode="anchor")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

results = collection.aggregate([
    {
        "$group": {
            "_id": "$team",
            "nb": {"$sum": {"$cond": [{"$gte": ["$pts", 20]}, 1, 0]}}
        }
    }
])

categories = []
counts = []

for result in results:
    categories.append(result["_id"])
    counts.append(result["nb"])

# Cr√©ation d'un barplot
plt.figure(figsize=(12, 6))
plt.bar(categories, counts, color='blue')
plt.title("Barplot des joueurs ayant plus de 20pts")
plt.xlabel("player")
plt.ylabel("points + passe d√©cisive + rebond")
plt.xticks(rotation=45, ha="right", rotation_mode="anchor")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

cur = list(collection.find())
cur

minj = []
pts = []

for result in cur:
    minj.append(result["minutes"])
    pts.append(result["pts"])

# Cr√©ation nuage de points
plt.scatter(minj, pts, color='blue')
plt.title("Temps de jeu vs. Points pour chaque joueur")
plt.xlabel("Temps de jeu")
plt.ylabel("Points")
plt.show()
```
